### Do you need to remap any fields?

```python
#Create dummy variables for the type of guns used in the shooting
firearms_data["ar_used"] = np.where((firearms_data["Classification"] == 3) & 
                                    (firearms_data["Used in Shooting?"] == 1), 1, 0)

firearms_data["rifle_used"] = np.where((firearms_data["Classification"] == 2) & 
                                    (firearms_data["Used in Shooting?"] == 1), 1, 0)

firearms_data["shotgun_used"] = np.where((firearms_data["Classification"] == 1) & 
                                    (firearms_data["Used in Shooting?"] == 1), 1, 0)

firearms_data["handgun_used"] = np.where((firearms_data["Classification"] == 0) & 
                                    (firearms_data["Used in Shooting?"] == 1), 1, 0)
```

```python
firearms_data.head()
```

```python
firearms_data = firearms_data.drop(["Classification"], axis=1)
```

```python
#Sum the gun used dummy variables
firearms_data = firearms_data.groupby('Case #').sum().reset_index()
```

```python
firearms_data
```

```python
#form lists in the case # column to account for commas
firearms_data = firearms_data.assign(**{'Case #':firearms_data['Case #'].str.split(',')})
```

```python
#Separate rows with multiple cases into rows
firearms_data = firearms_data.explode('Case #')
firearms_data
```

```python
#Change the cases with multiple case #'s to floats in order to sort'
#firearms_data['Case #'] = firearms_data['Case #'].apply(lambda x: x.replace(', ', '.'))
```

```python
firearms_data['Case #'] = firearms_data['Case #'].astype(float)
```

```python
#ascending order by case #
firearms_data.sort_values(by=['Case #'], inplace=True, ignore_index=True)
firearms_data
```

```python
#In order to merge, do the same to full data
full_data = full_data.assign(**{'Case #':full_data['Case #'].str.split(',')})

full_data = full_data.explode('Case #')

full_data['Case #'] = full_data['Case #'].astype(float)

full_data.sort_values(by=['Case #'], inplace=True, ignore_index=True)

full_data
```

```python
#Merge firearms data into full data
merge1 = pd.merge(full_data, firearms_data, left_on='Case #', right_on='Case #')
merge1
```

```python
#In order to merge community data into full data, the same case # alterations need to occur
community_data = community_data.assign(**{'Case #':community_data['Case #'].str.split('/')})

community_data = community_data.explode('Case #')

community_data['Case #'] = community_data['Case #'].astype(float)

community_data.sort_values(by=['Case #'], inplace=True, ignore_index=True)

community_data
```

```python
#Merge community data into the dataframe
merged_df = pd.merge(merge1, community_data, how='left', on='Case #')
merged_df
```

```python
#Convert variables that should be continuous 
merged_df['Total Population'] = merged_df['Total Population'].str.replace(',', '').astype(float)
merged_df['Size of Police Dept'] = merged_df['Size of Police Dept'].str.replace(',', '').astype(float)
merged_df.dtypes
```

```python
#Keep only continuous variables for standardization
con = merged_df.select_dtypes(include=['float64','int64']).columns
```

```python
con = merged_df[['Case #', 'Day', 'Month', 'Year', 'State Code', 'Region',
       'Multiple Locations', 'Other Location', 'Armed Person on Scene',
       'Number Killed', 'Number Injured', 'Used in Shooting?', 'ar_used',
       'rifle_used', 'shotgun_used', 'handgun_used', 'Location (Code)',
       'Time of Day', 'Total Population', '% Employed',
       '% High School Graduate', '% College Graduate',
       'Nearest Hospital (Miles)', 'N Mental Health Providers in Zip Code',
       'N Gun Stores in Zip Code', 'Size of Police Dept', 'Homicide Rate']]
con
```

```python
#Standardize dataframe
z_scores = (con-con.mean())/con.std()
z_scores
```

```python
z_scores[(z_scores > 3) | (z_scores < -3)] = 9999999999999

z_scores[(z_scores <= 3) | (z_scores >= -3)] = 0
```

```python
z_scores
```

```python
#There seems to be no outliers in the dataset

col = z_scores.columns

for i in col: 
    print(z_scores[i].value_counts())
```
